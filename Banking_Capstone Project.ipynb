The DecisionTreeClassifier class is instantiated with various hyperparameters specified within the parentheses.
class_weight=None sets the class weights to None, which means all classes are treated equally.
criterion='gini' uses the Gini impurity as the criterion to measure the quality of a split.
max_depth=5 sets the maximum depth of the decision tree to 5 levels.
max_features=None considers all features for splitting at each node.
max_leaf_nodes=20 limits the maximum number of leaf nodes to 20.
min_samples_leaf=1 sets the minimum number of samples required to be at a leaf node to 1.
min_samples_split=20 sets the minimum number of samples required to split an internal node to 20.
min_weight_fraction_leaf=0.0 sets the minimum weighted fraction of the sum of weights required to be at a leaf node to 0.0.
random_state=0 sets the random seed for reproducibility.
